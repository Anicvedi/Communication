```python
# Exercise 1: Sampling and Visualization
# Write Python code to sample a continuous signal with a given frequency at a specified sampling rate.
# Visualize the sampled signal alongside the continuous signal using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
print("Hello User")
print("This is the answer to my Exercise 1: Sampling and Visualization\n\
Write Python code to sample a continuous signal with a given frequency at a specified sampling rate.\n\
Visualize the sampled signal alongside the continuous signal using Matplotlib.")
f = 5 #int(input("Enter the signal frequency"))
fs = 50 #int(input("Enter sampling frequency"))
t = np.linspace(0,1,1000)
x = np.sin(2*np.pi*f*t)
print("Your entered signal is:\n")
plt.plot(t,x, label = 'Recieved Signal')
plt.title("Recieved Signal")
plt.grid()
```

```python
ts = np.arange(0,1,(1/fs))
xs = np.sin(2*np.pi*f*ts)
print("Your Sampled signal is:\n")
plt.stem(ts,xs, markerfmt='rx', basefmt='b-')
plt.title("Sampled Signal")
plt.grid()
```
